```python
import pandas as pd
from sqlalchemy import create_engine

# Connect to SQLite database
engine = create_engine("sqlite:///superstore.db")

print("Connected to SQLite database successfully!")

# 1. Display first 5 records
query1 = """
SELECT *
FROM sales
LIMIT 5;
"""
print("\nFirst 5 Records:")
print(pd.read_sql(query1, engine))

# 2. Category-wise Total Sales
query2 = """
SELECT Category,
SUM(Sales) AS TotalSales
FROM sales
GROUP BY Category;
"""
print("\nCategory-wise Total Sales:")
print(pd.read_sql(query2, engine))

# 3. Sales Greater Than Average
query3 = """
SELECT *
FROM sales
WHERE Sales >
(
    SELECT AVG(Sales)
    FROM sales
);
"""
print("\nSales Greater Than Average:")
print(pd.read_sql(query3, engine))

# 4. Customer-wise Total Sales (CTE)
query4 = """
WITH CustomerSales AS
(
    SELECT "Customer Name",
           SUM(Sales) AS TotalSales
    FROM sales
    GROUP BY "Customer Name"
)
SELECT *
FROM CustomerSales
ORDER BY TotalSales DESC;
"""
print("\nCustomer-wise Total Sales:")
print(pd.read_sql(query4, engine))

# 5. Category-wise Profit
query5 = """
SELECT Category,
SUM(Profit) AS TotalProfit
FROM sales
GROUP BY Category;
"""
print("\nCategory-wise Profit:")
print(pd.read_sql(query5, engine))

# 6. Category-wise Quantity
query6 = """
SELECT Category,
SUM(Quantity) AS TotalQuantity
FROM sales
GROUP BY Category;
"""
print("\nCategory-wise Quantity:")
print(pd.read_sql(query6, engine))

print("\nTask 2 Completed Successfully!")
```
